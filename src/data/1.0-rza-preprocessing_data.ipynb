## Preprocess Nifti images used for training, validation and testing

A script that is only used once, which loops over the whole BIDS folder strcuture, and saves all files in fnio (fast numpy io) format in the same folders with the same name (just the file ending is replaced by .fnio).

```python
from bids_dataset import create_bids_path_list_of_dicts, convert_nifti_to_fnio, contrasts
from pathlib import Path
```

```python
bids_root_path = '/media/DATA/farid_ma/dev/multiclass_softseg/MulticlassSoftSeg/data/external/ASNR-MICCAI-BraTS2023-GLI-Challenge/ASNR-MICCAI-BraTS2023-GLI-Challenge-TrainingData'
```

```python
list_of_path_dicts = create_bids_path_list_of_dicts(bids_root_path)
```

```python
print(list_of_path_dicts[0])
```

```python
print(bids_root_path)
```

```python
print(contrasts)
```

```python
for subject_dict in list_of_path_dicts:
        for contrast in contrasts:
            filepath = subject_dict.get(contrast)
            print(f"subject: {subject_dict.get('subject')} and contrast: {contrast}")
            #print(filepath)
            if filepath and Path(filepath).exists():  # Ensure the file exists
                path_as_string = str(filepath)
                convert_nifti_to_fnio(path_as_string)
                
        print(f"subject {subject_dict.get('subject')} done.")
        print(f"-------------------------------")
                
```
